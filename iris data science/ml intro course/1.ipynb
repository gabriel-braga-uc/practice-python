```python
v2 = [7, 5, 3, 5, 16]
v3= [4, 5, 6, 11, 1]
print("Vetores:", v2, v3)

#Soma de vetores
soma_vetores = v2[0] + v3[0], v2[1] + v3[1], v2[2] + v3[2], v2[3] + v3[3], v2[4] + v3[4]
print("Soma:", soma_vetores)

#Subtração de vetores
subtracao_vetores = v2[0] - v3[0], v2[1] - v3[1], v2[2] - v3[2], v2[3] - v3[3], v2[4] - v3[4]
print("Subtração:", subtracao_vetores)

#Norma de Vetores
norma_vetor_2 = (v2[0]**2 + v2[1]**2 + v2[2]**2 + v2[3]**2 + v2[4]**2)**0.5
print("Norma do vetor 2:", norma_vetor_2)

norma_vetor_3 = (v3[0]**2 + v3[1]**2 + v3[2]**2 + v3[3]**2 + v3[4]**2)**0.5
print("Norma do vetor 3:", norma_vetor_3)

#Angulo Entre Vetores
angulo_entre_vetores = np.arccos((v2[0]*v3[0] + v2[1]*v3[1] + v2[2]*v3[2] + v2[3]*v3[3] + v2[4]*v3[4])/(norma_vetor_2*norma_vetor_3))
print("Angulo entre os vetores radianos:", angulo_entre_vetores)
print("Angulo entre os vetores graus:", angulo_entre_vetores*180/np.pi)
```
